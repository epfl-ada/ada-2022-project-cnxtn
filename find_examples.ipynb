```python
import gzip
import pandas as pd
from tqdm import tqdm
import xml.etree.ElementTree as ET
from xml.dom import minidom
```

```python
plot_summaries = pd.read_csv(
    'MovieSummaries/plot_summaries.txt',
    delimiter='\t',
    names=['movie_id', 'plot']
)
```

```python
# parse an xml file by name
file = minidom.parse('generated_character_words_prettyprint_per_movie.xml')
```

```python
# retrieve characters
characters = file.getElementsByTagName('character')
```

```python
characters_list = []
for idx, character in enumerate(tqdm(characters)):

    words_list = []

    for word in character.getElementsByTagName('word'):
        attr = word.getAttribute('name')
        lemma = word.firstChild.data.lower()

        words_list.append((attr, lemma))

    name = character.getElementsByTagName('name').item(0).firstChild.data

    movie_id = int(character.getElementsByTagName('movie_id').item(0).firstChild.data.lower())

    characters_list.append((name, movie_id, words_list))
```

```python
def print_characters_lists(movie_id, characters_list):   
    filtered = list(filter(lambda x: x[1] == movie_id, characters_list)) #31186339
    for character in filtered:
        print("name: ", character[0])
        for attr in character[2]:
            print(attr[0],': ',attr[1])
        print(" ")
```
