```python
with gzip.open('corenlp_plot_summaries/3217.xml.gz', 'rt', encoding='utf-8') as f:
    root = ET.fromstring(f.read())

for child in root:
    print(child.tag, child.attrib)

for elem in root.iter('dep'):
    # print(elem.tag, elem.attrib, elem.text)
    if (elem.attrib['type'] == 'nsubj'):
        for child in elem:
            print(child.tag, child.attrib, child.text)
```

```python
for coreference in root[0][1]:
    print('Coreference: ')
    for mention in coreference:
        if(mention.attrib and (mention.attrib['representative'] == 'true')):
            print('This is the representative mention')
        for element in mention: 
            print(element.tag, element.attrib, element.text)
```

```python
import xmltodict

# Parse the XML file
with gzip.open('corenlp_plot_summaries/3217.xml.gz', 'rt', encoding='utf-8') as f:
    doc = xmltodict.parse(f.read())

# Iterate over the sentences in the document
for sentence in doc['root']['document']['sentences']['sentence']:
    # Iterate over the tokens in each sentence
    for token in sentence['tokens']['token']:
        # Extract the named entity tag and entity text
        ne = token['NER']
        word = token['word']
        if ((ne != 'O')): # and (ne == 'PERSON')):
            print(ne, word)
```

```python
for sentence in doc['root']['document']['sentences']['sentence']:
    if (int(sentence['@id']) == 15):
        for element in sentence['tokens']['token']:    #['basic-dependencies']['dep']:
            word = element['word']
            lemma = element['lemma']
            token_id = element['@id']
            if(int(token_id) in range(4,7)): 
                print(token_id, word, lemma)
```

```python
for coreference in doc['root']['document']['coreference']['coreference']:
    for mention in coreference['mention']:
        # rep = mention['@representative']
        sentence_id = int(mention['sentence'])
        start = int(mention['start'])
        end = int(mention['end'])
        head = int(mention['head'])
        print('mention in sentence ', sentence_id, ' starting at ', start, ' and ending in ', end)
        for element in doc['root']['document']['sentences']['sentence'][sentence_id-1]['tokens']['token']:
            if(int(element['@id']) in range(start,end)): 
                print(element['word'])
```
